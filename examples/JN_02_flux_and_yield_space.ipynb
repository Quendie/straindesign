Import cobrapy and straindesigner packages and load e_coli_core (textbook) model from cobra

```python
from pathlib import Path
import cobra
import straindesign as sd
model_path = (Path(cobra.__path__[0]) / "data" / "textbook.xml.gz")
model = cobra.io.read_sbml_model(str(model_path.resolve()))
```

Plot yield space

```python
%autoreload
# import matplotlib.pyplot as plt
y1 = ('Biomass_Ecoli_core','-EX_glc__D_e')
y2 = ('EX_etoh_e','-EX_glc__D_e')
y3 = ('ATPM','-EX_glc__D_e')

plot1 = sd.yield_space(model,(y1,y2))

plot2 = sd.yield_space(model,(y1,y2,y3))

plot3 = sd.phase_plane(model,('Biomass_Ecoli_core','EX_etoh_e'))

plot4 = sd.phase_plane(model,('Biomass_Ecoli_core','EX_etoh_e','ATPM'))
```
